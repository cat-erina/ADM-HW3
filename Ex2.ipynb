```python
import requests
import pandas as pd
import time
from bs4 import BeautifulSoup
import os
import csv
import itertools
import re
import csv
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
Dict = {}
ps = PorterStemmer()
attr = ['Intro', 'Plot']
stop = set(stopwords.words('english'))
```

```python
path = '/Users/yves/Desktop/Data_Science/first_year/first_semester/adm/adm_hw3/tsv_files/'

attr = ['Intro', 'Plot']

for filename in os.listdir(path):
    if filename.endswith(".tsv"):
        with open(path+filename) as fd:
            rd = csv.reader(fd, delimiter="\t", quotechar='"')
            L=[]
            for row in rd:
                if row[0] in attr:
                    row_cl = row[1].strip()  #  strips all kinds of trailing whitespace by default
                    row_cl_noPunct = re.sub(r'[^\w\s]','',row_cl) # remove puntuation
                    clean_row = [i for i in row_cl_noPunct.lower().split() if i not in stop] #remove stopwords
                    for word in clean_row:
                        L.append(ps.stem(word))
            final_words = list(set(L))
            for word in final_words:
                if word in Dict:
                    # append the new number to the existing array at this slot
                    Dict[word].append(filename)
                else:
                    # create a new array in this slot
                    Dict[word] = [filename]
            print(filename, ' Added')
print('Done')
```

```python
keys = ['disney', 'movies']
L = []
for idx in range(len(keys)):
    word = ps.stem(keys[idx])
    doc = Dict.get(word)
    if doc is None or len(doc) == 0:
        continue
    else:
        L.append(doc)
    print(len(L[idx]))

result = set(L[0]).intersection(*L)
```

```python
to_df = []
for movie in result:
    element_in_movie = {}
    with open(path+movie) as fd:
        rd = csv.reader(fd, delimiter="\t", quotechar='"')
        for row in rd:
            if row[0] == 'Name':
                element_in_movie['Name'] = row[1]
            elif row[0] == 'Intro':
                element_in_movie['Intro'] = row[1]
            elif row[0] == 'Url':
                element_in_movie['Url'] = row[1]
    to_df.append(element_in_movie)
df = pd.DataFrame(to_df)
df = df.reindex(('Name', 'Intro', 'Url'), axis=1)
df
```
